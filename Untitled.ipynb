```python
import pandas as pd
import numpy as np

%matplotlib inline

import seaborn as sns
import pylab as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15, 10))

sns.set(style='white')

mask=np.triu(np.ones_like(data.corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)


sns.heatmap(data.corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          annot=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});
```

```python
data=data.assign(Precio=1)
data
```

```python
precio=[]

for i in data.gas_type:
    if i == 'E10':
        precio.append(1.38)
    elif i == 'SP98':
        precio.append(1.46)
precio[:5]
```

```python
data.Precio=precio
data
```

```python
plt.figure(figsize=(15, 10))

sns.set(style='white')

mask=np.triu(np.ones_like(data.corr(), dtype=bool))

cmap=sns.diverging_palette(0, 10, as_cmap=True)


sns.heatmap(data.corr(),
           mask=mask,
          cmap=cmap,
          center=0,
          square=True,
          annot=True,
          linewidths=0.5,
          cbar_kws={'shrink': 0.5});
```

```python
data=data.assign(consumo_litro=data.distance*data.consume/100)
data
```

```python
data=data.assign(precio_litro=data.Precio*data.consumo_litro)
data
```

```python
data.groupby('gas_type').mean()
```

El tipo E10 tiene un consumo por litro mayor y un precio menor, pero la media consumida y precio es mayor que el SP98.
